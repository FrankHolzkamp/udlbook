```python
y = np.random.normal(1, 0.8, 1000)
# print(y)
# print(y.size)
x = np.arange(0,100,0.1)
print(x)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate mean and standard deviation of y
mu, std = np.mean(y), np.std(y)
print(mu)
print(std)

# Create a range of x values for the bell curve
x_curve = np.linspace(min(y), max(y), 100)
delta_x = x_curve[1] - x_curve[0]

# Calculate the PDF for the normal distribution
pdf = norm.pdf(x_curve, mu, std)

# Calculate surprisal
surprisal = np.log(1/pdf)
# entropy=np.average(surprisal)
entropy_approx = np.sum(pdf * surprisal * delta_x)
print(entropy_approx)
entropy_approx_from_y = np.sum(y * np.log(1/y))
print(entropy_approx_from_y)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the histogram of y on the first y-axis
ax1.hist(y, bins=50, density=True, alpha=0.6, color='g', label='Histogram of y')

# Plot the bell curve (PDF) on the first y-axis
ax1.plot(x_curve, pdf, 'r', linewidth=2, label='Bell Curve (Normal Distribution)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density', color='g')
ax1.tick_params(axis='y', labelcolor='g')

# Create a second y-axis for surprisal
ax2 = ax1.twinx()
ax2.plot(x_curve, surprisal, 'b--', linewidth=2, label='Surprisal (-log PDF)')
ax2.set_ylabel('Surprisal', color='b')
ax2.tick_params(axis='y', labelcolor='b')
ax2.invert_yaxis() # Often surprisal is inverted to show 'less surprising' as higher

plt.title('Bell Curve (Normal Distribution) and Surprisal vs. Histogram of y')

# Combine legends from both axes
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

ax1.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values, avoiding x=0 as log is undefined at 0
x_log = np.linspace(0.01, 5, 400) # Start slightly above 0

# Calculate natural logarithm (base e)
y_ln = np.log(x_log)

# Calculate base-10 logarithm
y_log10 = np.log10(x_log)

plt.figure(figsize=(10, 6))

# Plot natural logarithm
plt.plot(x_log, y_ln, label='ln(x) (Natural Logarithm)', color='blue')

# Plot base-10 logarithm
plt.plot(x_log, y_log10, label='log10(x) (Base 10 Logarithm)', color='red', linestyle='--')

# Add vertical asymptote line at x=0
plt.axvline(x=0, color='gray', linestyle=':', label='Vertical Asymptote at x=0')

# Add horizontal line at y=0 (where log(1)=0)
plt.axhline(y=0, color='black', linestyle=':', label='log(1)=0')

# Annotations for characteristics
plt.text(0.5, -2, 'Log is undefined for x <= 0', color='purple', fontsize=10, ha='center')
plt.text(2.5, 1.5, 'Increasing function', color='green', fontsize=10)

plt.title('Important Characteristics of the Logarithm Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.grid(True)
plt.legend()
plt.ylim(-5, 2) # Adjust y-limits for better visualization
plt.show()
```
